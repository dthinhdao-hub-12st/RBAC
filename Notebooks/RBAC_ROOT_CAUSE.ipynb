## ID = 11

```python
df_unplanned_11 = df_unplanned[df_unplanned['ID'] == '11']
df_unplanned_11
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# --- 1. Tạo cột đánh dấu lỗi ID=11 ---
df['IS_ERROR_11'] = 0
df.loc[df.index.isin(df_unplanned_11.index), 'IS_ERROR_11'] = 1

# --- 2. Hàm tính Cramér’s V để đo mức độ ảnh hưởng (0 → 1) ---
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(r - 1, k - 1))))

# --- 3. Hàm kiểm định Chi-square + Cramér’s V ---
def chi_square_test(df, column):
    # Chuẩn hóa: loại NaN & chuyển về category
    data = df[[column, 'IS_ERROR_11']].dropna()
    data[column] = data[column].astype('category')

    # Tạo bảng chéo
    contingency_table = pd.crosstab(data[column], data['IS_ERROR_11'])

    # Nếu bảng quá nhỏ hoặc chỉ có 1 chiều thì bỏ qua
    if contingency_table.shape[0] < 2 or contingency_table.shape[1] < 2:
        print(f"\n Không đủ dữ liệu để kiểm định với {column}")
        return

    # Chi-square
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Cramér’s V
    V = cramers_v(contingency_table)

    print(f"\n Yếu tố: {column}")
    print(f"  Chi-square = {chi2:.3f}")
    print(f"  p-value    = {p:.5f}")
    print(f"  Cramér's V = {V:.4f}")

    if p < 0.05:
        print("    Kết luận: Có ảnh hưởng đáng kể đến lỗi ID = 11")
    else:
        print("    Kết luận: Không ảnh hưởng đáng kể")

# --- 4. Chạy kiểm định cho các yếu tố ---
for col in ['LINE_NAME', 'SIZE_TYPE', 'CREW_ID']:
    chi_square_test(df, col)
```

Cả LINE_NAME, SIZE_TYPE, và CREW_ID đều có ảnh hưởng đáng kể tới lỗi ID = 11 (p-value = 0.00000).

Tuy nhiên, mức độ ảnh hưởng khác nhau rõ rệt dựa trên Cramér’s V:

- LINE_NAME là yếu tố ảnh hưởng mạnh nhất (V = 0.2226) → Đây là yếu tố chính cần tập trung.

- SIZE_TYPE có ảnh hưởng nhưng yếu hơn (V = 0.1796).

- CREW_ID ảnh hưởng rất nhỏ, gần như không đáng kể (V = 0.0314)
