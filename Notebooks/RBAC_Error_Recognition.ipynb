```python
import pandas as pd
from scipy.stats import chi2_contingency

df['IS_ERROR_11'] = 0

# Gán = 1 nếu bản ghi nằm trong df_unplanned_11 (cùng index)
df.loc[df.index.isin(df_unplanned_11.index), 'IS_ERROR_11'] = 1

def chi_square_test(df, column):
    contingency_table = pd.crosstab(df[column], df['IS_ERROR_11'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"\n Yếu tố: {column}")
    print("Chi-square:", chi2)
    print("p-value:", p)
    if p < 0.05:
        print("Kết luận: Có ảnh hưởng đáng kể đến lỗi ID = 11")
    else:
        print("Kết luận: Không ảnh hưởng đáng kể")

# Chạy kiểm định:
for col in ['LINE_NAME', 'SIZE_TYPE', 'CREW_ID']:
    chi_square_test(df, col)
```

```python
def detect_anomaly_full(df, cluster_name, group_col, top_n=5):
    df_cluster = df[df['cluster_label'] == cluster_name]
    counts = df_cluster[group_col].value_counts()
    percent = df_cluster[group_col].value_counts(normalize=True) * 100

    result = pd.DataFrame({
        'count': counts,
        'percent (%)': percent
    }).head(top_n)

    return result
```

```python
for cluster in df_unplanned_mh['cluster_label'].unique():
    print(f"\n======  Cluster: {cluster} ======")
    anomaly_line = detect_anomaly_full(df_unplanned_mh, cluster, "LINE_NAME")
    anomaly_crew = detect_anomaly_full(df_unplanned_mh, cluster, "CREW_ID")
    anomaly_size = detect_anomaly_full(df_unplanned_mh, cluster, "SIZE_TYPE")

    print("\n➡ Top 5 LINE_NAME bất thường:")
    print(anomaly_line)

    print("\n➡ Top 5 CREW_ID bất thường:")
    print(anomaly_crew)

    print("\n➡ Top 5 SIZE_TYPE bất thường:")
    print(anomaly_size)
```

```python
def detect_high_impact_low_frequency(df, error_col='Error_Type', downtime_col='DOWNTIME'):
    # Tổng toàn bộ lỗi và downtime
    total_errors = df.shape[0]
    total_downtime = df[downtime_col].sum()

    # Gom nhóm theo từng loại lỗi
    summary = df.groupby(error_col).agg(
        error_count=(error_col, 'count'),
        total_downtime=(downtime_col, 'sum')
    )

    # Tính % lỗi và % downtime
    summary['error_percent'] = summary['error_count'] / total_errors * 100
    summary['downtime_percent'] = summary['total_downtime'] / total_downtime * 100

    # Lọc lỗi xuất hiện ít (<30%) nhưng gây downtime lớn (>70%)
    result = summary[(summary['error_percent'] < 30) & (summary['downtime_percent'] > 70)]

    return summary.sort_values(by='downtime_percent', ascending=False), result

# Nếu cột lỗi của bạn tên là "cluster_label"
summary, high_impact_errors = detect_high_impact_low_frequency(
    df_unplanned_mh, error_col='cluster_label', downtime_col='DOWNTIME'
)

print("Tổng quan lỗi & downtime theo từng loại:")
print(summary)

print("\nLỖI ÍT NHƯNG GÂY ẢNH HƯỞNG LỚN:")
print(high_impact_errors if not high_impact_errors.empty else "Không tìm thấy loại lỗi nào thuộc dạng này.")
```
